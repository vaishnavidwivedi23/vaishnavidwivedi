```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the data
data = pd.read_excel("C:\\Users\\Flash\\Desktop\\Projects\\Whether Analysis Report\\weather_data.xlsx")
print(" Data loaded successfully!\n")
print(data.head())
```

```python
# Step 2: Basic Info
print("\n Data Summary:")
print(data.describe())
```

```python
# Step 3: Average Temperature per Month (overall)
avg_temp = data.groupby("Month")["Temperature"].mean()
print("\nðŸŒ¡ Average Temperature per Month:\n", avg_temp)
```

```python
# Step 4: Average Rainfall per City
avg_rainfall_city = data.groupby("City")["Rainfall"].mean()
print("\nâ˜” Average Rainfall per City:\n", avg_rainfall_city)
```

```python
# Step 5: Compare Two Cities (Delhi vs Mumbai)
city_compare = data[data["City"].isin(["Delhi", "Mumbai"])]
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=city_compare, x="Month", y="Temperature", hue="City", marker="o")
plt.title("Monthly Temperature Comparison: Delhi vs Mumbai")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.show()
```

```python
# Step 6: Rainfall Pattern
plt.figure(figsize=(10, 5))
sns.barplot(data=data, x="Month", y="Rainfall", hue="City")
plt.title("Monthly Rainfall Pattern by City")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
# Step 7: Correlation Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(data[["Temperature", "Humidity", "Rainfall"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation between Weather Factors")
plt.show()
```

```python
# Step 8: Findings
best_city = avg_rainfall_city.idxmax()
print(f"\n City with Highest Average Rainfall: {best_city}")
```

```python
hottest_month = avg_temp.idxmax()
print(f" Hottest Month on Average: {hottest_month}")
```
